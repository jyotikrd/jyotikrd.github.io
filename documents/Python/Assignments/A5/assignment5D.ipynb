Keep asking the user to enter numbers until the user decides to stop by pressing any alphabet key. The first number should either be `1` or `2` which needs to be stored as `displayChoice`. Remaining values should be stored as a list.

Write a function that takes this list of numbers and `displayChoice` as parameters. 

The function should compute and return `max`, `min`, `sum`, `count` and `average` of numbers in the list based on following values of `displayChoice`

*   If `displayChoice` is `1`, return above as a single list and print as single list
*   If `displayChoice` is `2`, return above as tuple and print as separate values

**Note:** 

*   User should be asked to enter the input on new line one value at a time
*   No restrictions on type of numbers that will form the list
*   Use inbuilt function to check for alphabet entry
*   Display the Average values upto three decimal places




---


**Example 1**

Input

```
2
2
10
64
32
54
g
```

Output


```
max: 64
min: 2
sum: 162
count: 5
average: 32.400
```

Input

```
1
2

```

Output


```
[max, min, sum, count, average]
[64, 2, 162, 5, 32.400]
```
