### Boolean Operators

Programming languages give us access to operators that can compare something. That might not always be enough, sometimes we might need to combine comparisions, or know whether one or all of many comparisions are True. Thus, we need something that will work with Boolean values 

To help us with that, Python lets us use the following operators to work with Boolean values:

| Operator | Usage   | Result                                                 |
|----------|---------|--------------------------------------------------------|
| and      | `A and B` | True if both A & B are True, False otherwise           |
| or       | `A or B`  | True if atleast one of A / B is true, False otherwise  |
| not      | `not A`   | "Negation": True if A is False, False if A is True     |

### Complex Conditions

Often in your programs, you'll find a need to 'chain' multiple conditions, or execute a block of code if one of many _different_ conditiosn are met. 

In that case, you can combine the knowledge of Boolean Operators and conditionals to write complex conditions

```
# Problem: Let's say we can either cross the Thames river using the two bridges we talked about
# before, or use the new ShiP that has just launched. But the latter costs money.
# We can try and represent this in terms of boolean values and operators

bridge_1_down = True
bridge_2_down = False

ship_cost = 5
money_in_wallet = 7

can_cross = (bridge_1_down and bridge_2_down) or (money_in_wallet >= ship_cost)
# Explanation: We can cross if both bridges have their bascules down OR if we have enough money
# for the ShiP.

print(can_cross)

# Play with adifferent values of all the four variables above to see the final result
```

### Branching

The ability to make decisions based on a condition is one of the core abilities of a program or let's say LIFE.

* In an inventory-tracking program, you might want to block all purchases and send a notification to the owner once the stock of an item reaches zero.
* In a stock trading application, you might want to automatically sell the stock when the value goes above / below a given value.

And many other instances. Python gives us the ability to make such decisions, called _Branching_ or _Conditionals_. The most simple example of Conditionals is the `if` block.


```
# Going back to the lecture, consider our dog, Fluffy, that wanted to cross over those
# two bridges.
# We want to print out a message if it can pass the bridge.

# Try changing these values and run the block again to see how the output changes.
bridge_1_down = True
bridge_2_down = False

# We need both bridges to be down for Fluffy to cross Thames.
can_cross = bridge_1_down and bridge_2_down

if can_cross:
  print("Fluffy can cross the Thames! ðŸ™‚")

# Note the indentation of the print statements. That signifies that the statements
# are inside a 'block' of code, which here is an 'if' block
```

Let's say we have different things to do when a condition is met vs when it _isn't_, then we can use the `else` block. 

```
# We want to print out a message with ðŸ™‚ if it can cross the Thames else other with ðŸ˜­ if it cannot.

# Try changing these values and run the block again to see how the output changes.
bridge_1_down = True
bridge_2_down = False

# We need both bridges to be down for Fluffy to cross Thames.
can_cross = bridge_1_down and bridge_2_down

if can_cross:
  print("Fluffy can cross the Thames! ðŸ™‚")
else:
  print("Fluffy can not cross the Thames! ðŸ˜­ ")
```

There might be cases where if a given condition isn't met, you might want to check another condition and run some code if the latter condition is `True`. For such a case, we have the `elif` block.
This is a short form of the phrase 'else if'. 

```
# Problem: Going back to the lecture, consider our dog, Fluffy, that wanted to cross over those
# two bridges. 
# We want to print out a message in the following cases:
#   * If Fluffy crossed the Thames!
#   * If Fluffy can cross Half-way with the FIRST bridge
#   * If Fluffy can not cross the Thames and is stuck on other side

# Try changing these values and run the block again to see how the output changes.
bridge_1_down = True
bridge_2_down = False

# We need both bridges to be closed for Fluffy to cross the Thames.
can_cross = bridge_1_down and bridge_2_down
 
if can_cross:
  print("Fluffy crossed the Thames! ðŸ™‚")
elif bridge_1_down: # This runs if ONLY bridge 1 is down. Can you tell how?
  print("Fluffy made it half-way over Thames! ðŸ™Œ")
else:
  print("Fluffy can not cross the Thames and is stuck on other side! ðŸ˜­")
```

You can also have multiple `elif` blocks following an `if` block.

```
# Problem: Let's say we want to convert a numeric grade into a letter grade.
# This code block shows how we can do that with conditionals.

# Try changing this value and see how to output changes
grade = 92

if grade >= 90:
  print("You got an !")
elif grade >= 80: # This block checks if 80 <= grade < 90. Can you tell how?
  print("You got a B")
elif grade >= 70:
  print("You got a C")
else:
  print("Better luck next time!")
```

### Nested Conditionals

Sometimes you might need to check a condition _after_ you know another condition has already been met. Maybe you see that a customer has enough money for a product, but want to know if they have enough money for you to recommend a deal to them.

In such cases, you can _nest_ conditional blocks. They are just like normal conditional blocks, but inside another block. 

**Note:** As you can see below, the statements inside the nested conditional block are indented _twice_. As a rule, if a statement is supposed to be inside an _nth_ nested block, the statement should be indented _n_ times.

```
# Problem: We wrote code for converting a numeric grade -> letter grade. Now we
# might want more granularity in the letter grades with 'A+', 'B-' and so on.

# This code block shows how we can do that with nested conditionals.

grade = 82

if grade >= 90:
  if grade >= 95:
    print("You have an A+")
  else:
    print("You have an A-")
elif grade >= 80: 
  if grade >= 85:
    print("You have an B+")
  else:
    print("You have an B-")

# Note: Your conditonal blocks don't always have to end with `else` an `elif` block
# can be the last one too! But if an `else` block does exist, it need to be at the
# end. 
```

### Conditional Assignment
Sometimes you might have a branch that is essentially the same code with a different expression.  For cases like this, python has conditional expressions ( or assignment)

They can be thought of as a condensed form of a basic `if / else` block. Other languages have similar statements called _ternary statements_.

```
# Problem: Going back to our example of Fluffy trying to cross the Thames with two bridges in Series.
# Let's try and print out something for the cases where Fluffy can / can not cross
# the bridges, but using Conditional Expressions this time around.

bridge_1_down = True
bridge_2_down = False

# CASE A: Bridges in Series
can_cross = bridge_1_down and bridge_2_down

sentence_to_print = "Fluffy crossed the Thames! ðŸ™‚" if can_cross else "Fluffy can not cross the Thames! ðŸ˜­"
print(sentence_to_print)
```
