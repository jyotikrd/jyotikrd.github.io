# Tax Calculation

This assignment will help with better understanding of nested conditional statements and expression. 


You should accept four inputs from the user, namely,
*   Annual household income (in $)
*   status of filing ('s' for single, 'h' for head of household, 'j' for joint)
*   have children ('y' yes or 'n' no)
*   are you a student ('y' yes or 'n' no)

and accordingly calculate the tax rate and amount to be paid in taxes.

\begin{array}{ccc}
\text{Tax Rate}&\text{Single}&\text{Head of Household}&\text{Joint}\\
10\%&\$0+&\$0+&\$0+\\
15\%&\$10000+&\$15000+&\$20000+\\
20\%&\$40000+&\$53000+&\$80000+\\
25\%&\$85000+&\$85000+&\$172000+\\
30\%&\$165000+&\$165000+&\$327000+\\
\end{array}

**Deduction:**
1. Has Children: $-3.5\%$
2. Are Student: $-10\%$ if income is $\lt \$12000$ or $-5\%$ if income $\geq \$12000$

PS: This question is only for education purpose, don't use to calculate the actual amount.

Output Expected:
```
Annual Income: ___
Filing Status: ___
Children: ___(yes or no)
Student: ___(yes or no)
Tax Rate Calculated: ___
Tax Amount to be Paid: ___
```

Example Input:
```
Enter Annual Income: 11500.00
Enter Filing Status: s
Do you have any Children ? n
Are you a Student ? y
```

Example Output:
```
Annual Income: $11500.00
Filing Status: single
Children: no
Student: yes
Tax Rate Calculated: 5% 
Tax Amount to be Paid: $575.00
```
