# Perimeter and Area of a Square

In this activity you will learn, how to accept the input from the user regarding side of a square in Console and process that information to display the perimeter and area of square. A square is a flat shape, in one plane, defined by four points at the four corners. A square has four sides all of equal length, and four corners, all right angles (90 degree angles). A square is a kind of rectangle.

**Hint**:- Datatype conversion (casting) is required in order to calculate the area and perimeter. 

![alt text](https://drive.google.com/uc?id=1Yj5_nYBjXdwaOBx6nOU1551uPOxJ-4SI)

### Perimeter of square = 4 x side
### Area of Square = side x side

The desired output should be as follows **with 2 decimal places for every blank**:


```
Perimeter of square with side ______ cm is ________ cm
Area of square with side ______ cm is ________ square cm
```
