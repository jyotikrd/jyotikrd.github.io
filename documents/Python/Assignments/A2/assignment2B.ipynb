# Age Group Identifier

In this assignment you will learn about the basic of the if, else or elif. Suppose you are an airline company's employee, and are given a task to classify if your passengers flying with your airlines are

*   Infant(0-2 years)
*   Minor(3-18 years)
*   Adult(19-64 years)
*   Senior(65+ years)

Your program should accept the passengers first name, last name, reference number(can only be 6 digit), and age.

Hint:- For validating the reference number try using this concept

https://www.geeksforgeeks.org/python-string-length-len/

Input Example 1:
```
First Name: John
Last Name: Doe
Age: 45
Reference Number: FR53FY
```
OutputExample 1:
```
Doe, John with Reference Number FR53FY is Adult.
```


Input Example 2:
```
First Name: John
Last Name: Doe
Age: 45
Reference Number: DT4HF63
```
Output Example 2:
```
Wrong Input for Reference Number.
```
