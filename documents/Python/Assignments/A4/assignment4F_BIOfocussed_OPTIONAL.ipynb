# **BIO-1: Calculate melting point of the given DNA primer**
For sequences less than 14 nucleotides melting temperature is calculated by:


$Tm= (wA+xT) \times 2 + (yG+zC) \times 4$


*   Where, $wA$ - no. of Adenine in the molecule
*   $xT$ - no. of Thymines and so on...



For sequences greater than or equal to 14 nucleotides melting temperature is given by:

$Tm= 64.9 +41\times \frac{yG+zC-16.4}{wA+xT+yG+zC}$
