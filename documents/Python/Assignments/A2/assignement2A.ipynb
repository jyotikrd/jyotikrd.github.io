# SI or CI calculation based on User-choice

In this activity you will calculate Simple interest or Compound Interest depend on the user input. Your program should accept the users input for principal amount, interest rate, time period, and type of interest calculation(1 for simple or 2 for compound). If a user select Compound interest, they should accept another input from user regarding the interval. Calculate the appropriate interest amount.

![alt text](https://getcalc.com/formula/finance/si-simple-interest.png)
![alt text](https://drive.google.com/uc?id=1kdHwYtyoS6ACa9RHpHMDk_nLFJR7R9Lk)

Expected Output:
```
Type of Interest requested by user is: ____
Principal Amount: ____
Interest Rate: ____
Time Period: ____
Interest Calculated: ____
```

Example Input(Simple):
```
Principal Amount: 5000
Time Period: 2
Interest Rate: 5
Type of interest: 1
```
Example Output(Simple):
```
Type of Interest requested by user is: Simple
Principal Amount: $5000.00
Interest Rate: 5%
Time Period: 2
Interest Calculated: $500.00
```
Example Input(Compound):
```
Principal Amount: 5000
Time Period: 2
Interest Rate: 5
Type of interest: 2
Interval per year: 4
```
Example Output(Compound):
```
Type of Interest requested by user is: Compound
Principal Amount: $5000.00
Interest Rate: 5%
Time Period: 2
Interval: 4
Interest Calculated: $522.43
```
