# Squares of positive even integers 
Write a program to get input integer, `n`, from the user, then print out the squares of all the positive even integers that are smaller than `n` .

**Example 1:**
let n=15, then 2,4,6,8,10,12,14 are positive even integers less than 15. So, print square of these numbers i.e. 4, 16, 36, 64,..,196

`15`

`4 16 36 64 100 144 196`

**Example 2:**
let n=2, then there are NO positive even integers less than 2. So, print 'Not Possible'

`2`

`Not Possible`
