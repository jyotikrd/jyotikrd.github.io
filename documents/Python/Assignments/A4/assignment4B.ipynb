Now, This won't work like that. We cannot compare a list to an int.

What we want is a Boolean list `[True, False, False, True, False]`.



```
# Creating Boolean List

list1 = [1, 3, 5, 2, 4]
threshold = 3

truth_list = []

for x in list1:
  if x < threshold:
    truth_list.append(True)
  else:
    truth_list.append(False)

print(truth_list)
```
