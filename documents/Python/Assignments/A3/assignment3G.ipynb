# Blackbeard Treasure
Blackbeard, the most notorious pirate in the history stumbled across some treasure. Being the superstitious man he is, he believes in taking treasure money only in counts of "Crew Numbers".

 A Crew Number is a 3-digit number, where the sum of the first two digits of the number equals the 3rd digit of the number.

Unfortunately, Blackbeard only knows one Crew Number 235 where 2+3 gives 5 as third digit. Similary 256 is not a Crew Number as 2+5 does not equal to 6. 

Write a program that prints the maximum Crew Number possible below or equal to certain user-input integer.

Input 1
```
200
```
Output 1

```
189
```

Input 2
```
100
```
Output 2

```
sad
```

Input 3
```
500
```
Output 3

```
459
```
