# Circumference and Area of Circle

In this activity you will learn, how to accept the input from the user regarding radius of a circle in Console and process that information to display the circumference and area of circle. For this activity we will require to use the exponentional expression discussed in the class. 

You will have to use two variables, one for the user input and another for the constant pi (=22/7).

Hint:- Datatype conversion (casting) is required in order to calculate the area and perimeter. 


![alt text](https://qph.fs.quoracdn.net/main-qimg-ddee5c6376bdf437f64df57f722c2edd)

### Perimeter of Circle = 2 x pi x radius
### Area of Circle = pi x radius^2  (use ** as exponential)

The desired output should be as follows **with 2 decimal places for every blank**:


```
Circumference of circle with radius __ cm is ____ cm
Area of circle with radius __ cm  is __ square cm
```
