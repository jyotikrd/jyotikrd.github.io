````
import matplotlib.pyplot as plt

# Pie chart
grades = ['A', 'B', 'C', 'D', 'F']
majors = ['GRY', 'SLY', 'HUF', 'RAV', 'BEA', 'DUR', 'MUG']
programs = ['Freshman', 'Sophomore', 'Junior', 'Senior', 'Masters', 'PhD']


# EXAMPLE 
grades_distribution = [15, 5, 10, 4, 2] #Replace it with actual list from your code
majors_distribution = [10, 10, 10, 10, 10, 10,10] #Replace it with actual list from your code
programs_distribution = [10, 10, 10, 10, 10, 10] #Replace it with actual list from your code


# only "explode" the 1st slice (i.e. 'A')
explode = (0.1, 0, 0, 0, 0)

#add colors
# You may choose favorite color from https://www.color-hex.com/
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#f5f5dc' ]


fig1, ax1 = plt.subplots()
ax1.pie(grades_distribution, labels=grades, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90)

fig2, ax2 = plt.subplots()
ax2.pie(majors_distribution, labels=majors, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90)

fig3, ax3 = plt.subplots()
ax3.pie(programs_distribution, labels=programs, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90)



# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')
ax2.axis('equal')
ax3.axis('equal')
plt.tight_layout()
plt.show()
````
