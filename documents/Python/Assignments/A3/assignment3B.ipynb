# Factorial

A factorial of a number `n` , represented as n! is the product of all the numbers leading upto `n`. Note: $$0! = 1$$

**Example:** factorial of 4, represented as 4!= 4x3x2x1 = 24. Write a program that calculates factorial of number, `n`, given by the user.

Input

`4`

Output

`4! = 24`
