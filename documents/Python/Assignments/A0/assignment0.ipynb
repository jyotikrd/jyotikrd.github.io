# Indentation

*   `def` is **Level 0** indentation
*   `print("This is a calculator")` , `if`, and `elif` are **Level 1** (single *tab*) indentation within `def` block
*   `print("Sum = ",a+b)` is **Level 2** indentation (2 **tabs** from `def` block or single *tab* from `if` block)
*    `print("Difference = ",a-b)` is **Level 2** indentation (2 **tabs** from `def` block or single *tab* from `elif` block)
*    ` calculator(10, 12, '-')` is **Level 0** i.e. same level as `def ` block



## Calculator
 Calculating sum or difference of two numbers, *10*, and *12* based on value of *key*

```
# define a calculator function
def calculator(a, b, key):
  print("This is a calculator")

  if(key == '+'):  # if value of key is +, perform addition
    print("Sum = ",a+b)
  elif(key == '-'):           # else if value of key is -, perform subtraction
    print("Difference = ",a-b)

# call calculator function with values 10, 12 and -
calculator(10, 12, '-')
```

# Gross Pay
Write a Program (WAP) to calculate Gross Pay ($) from user-input values for number of hours worked and hourly pay rate


```
# We can use input() to directly ask user to give their choice of values 
# However, to make it user-friendly, we added String messages. For eg: "Enter the number of hours:  " within input()
# Unfortunately, the input is considered as String: sequence of character(s) i.e. '3' or '15.5' (example)
# So we will see how we use int() and float() to convert these to corresponding numeric values ,i.e. 3 and 15.5
# The numeric values will then be stored, respectively, in variables, hours and payRate

# 1. get the number of hours worked
hours = int(input("Enter the number of hours:  "))  # int() converts input from user to numeric (integer) type (eg: 3)

# 2. get the hourly payrate
payRate = float(input("Enter the hourly pay rate: "))  # float() converts input from user to numeric (floating:think real values
                                                       #  with decimals) type as payrate can include cents (eg: 15.5)

# 3. calculate grossPay
grossPay = hours * payRate

# 4. display the result
print("The gross pay is = $",grossPay)
```
