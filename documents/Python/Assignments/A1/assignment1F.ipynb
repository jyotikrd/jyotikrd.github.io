# Calculate the Compound Interest
In this activity you will learn how to calculate compound interest and use of different variables and expression. In this activity you are expected to take 4 different input from the user as follows


```
P	=	initial principal balance in $
r	=	interest rate (in percentage) eg: 5 , but for calculation your code needs to use it as 0.05
n	=	number of times interest applied per time period (like twice or 4 times a year)   (assume integer)
T	=	number of time periods elapsed  (assume integer)
```
# $${A = P (1+\frac{r}{n})^{nT}}$$

# $${CI = A - P} $$


Output expected (with 2 decimal places for each blank):


```
Compound interest calculated is $_______
Final amount calculated is $_______
```

**Example:**
If an amount of $5,000 is deposited into a savings account at an annual interest rate of 5%, compounded monthly, the value of the investment after 10 years can be calculated as follows...

P = 5000

r = 5/100 = 0.05 (decimal)

n = 12 (compunded monthly, so 12 times in a year)

T = 10


If we plug those figures into the formula, we get the following:

$$A = 5000 \times (1 + \frac{0.05}{12})^{(12 \times 10)} = 8235.05$$

```
Compound interest calculated is $3235.05
Final amount calculated is $8235.05
```
