### Repetitions / Looping

Often we need to repeat a block of code in our programs a given number of times, or until a specific condition is met:
* We might need to find all numbers from 1-100 that are divisible by a specific number
* We might want to run a game until the user enters an 'exit' command
* We often need to search for a given item from a group of items

For such use cases and more, Python gives us _loops_. An example from what we discussed in class is below:

````
# Problem: Let's take the example of eating cereal, as discussed in the lecture.
# This cell prints out the procedure one would take to eat cereal.

# Don't worry about the implementation for now, you'll learn how to do this by the end of class!

num_spoonfuls_of_cereal = 6

print("Add cereal to bowl.")
print("Add milk to bowl.")

for i in range(num_spoonfuls_of_cereal): # Our loop condition.
  # Our loop body.
  print(f"Eat a spoonful of cereal. Number of spoonfuls left: {num_spoonfuls_of_cereal - (i+1)}")

print("We're done with the cereal")
````

### While Loops

While loops run as long as the given _loop condition_ is true. Because of this, it's usually used when the number of repetitions is _unknown_. 

It also allows an `else` block at the end which is run once the condition is no longer True.

````
# Let's print the first 20 natural numbers using a while loop:

current_num = 1

while current_num <= 20:
  print(current_num)
  current_num += 1

# Note that at the end of the loop, the value of current_num will be 21
print("Loop has ended!")
print(f"Value of current_num is {current_num}")
````

**Be cautious of INFINITE LOOP**

````
# Let's print the first 20 natural numbers using a while loop:

current_num = 1

while current_num > 0:  # some poor mistake
  print(current_num)
  current_num += 1

# Note that at the end of the loop, the value of current_num will be 21
print("Loop has ended!")
print(f"Value of current_num is {current_num}")
````

````
# Problem: Let's say you have unlimited cereal, and the user can stop whenever they want.
# This cell uses a while loop to let the user eat cereal until they provide an 'exit' command

print("We've not unlimited cereal!")

user_command = "default_command"

while user_command != "exit": # Enter the condition here.
  print("You ate a spoonful of cereal")
  user_command = input("Enter 'exit' if you're done eating cereal. Enter anything else if you want more! : ")

print("We're done eating cereal!")
````

### For Loops

For loops in Python are used to _iterate_ over something (like a list of numbers or items). These are normally used when the number of repetitions is _known_. 

A basic use for these is to loops over a range of numbers, one at a time. We use the `range` function to get this list of numbers. It has to following usage:
```
range(<start_value>=0, <end_value>, <step>=1)
```

**Note:** The `range` function generates numbers starting from `start_value` until `end_value` (not including it) with incrementation of `step`

````
# default step is 1
x = range(1,6) # x is a range of values: (start)1,2,3,4,5 (not including end value i.e. 6)

# y is an iterator that will go over x, element by element
y = iter(x)

# next() function moves the iterator to next element in the range (x)
print(next(y))
print(next(y))
print(next(y))
print(next(y))
print(next(y))

# try doing another print(next(y))
````

````
# simplifying above repetitions of print by using for-loop

for y in range(1,6):
  print(y)
````

````
# deafuly values of start (0) and step (1)
for y in range(6):
  print(y)
````

````
# step = 2
for y in range(1,6,2):
  print(y)
````

````
# step = -1
for y in range(6,0,-1):
  print(y)
````

````
# Let's try printing out the first 20 whole numbers.

for i in range(20):
  # The variable `i` contains the number of the current iteration
  print(i)
````

````
# We can use the <step> parameter to change how big the increments are between
# two subsequent numbers that are generated.

# Using the step parameter gives us an easy way to generate even numbers.
# Problem: Print out all odd numbers from 40-70

for i in range(41, 70, 2):
  print(i, end=" ")
````

The variable `i` in the examples above (you can name it whatever we want) is the _loop variable_ and can be used as any other variable inside the loop. 

The only difference is that it will have a different value in each iteration of the loop.

````
# We can try printing out the squares of the first 15 natural numbers.

for i in range(1, 16): # Generate numbers from 1 to 15 (i.e. 16-1)
  # You can use the loop variable `i` like you would any other variable
  print(f"The square of {i} is {i**2}")
````

````
# How about drawing out a pyramid where we take its height as user input?
# Pyramid here means for every level (line), we print that many number of *
# If the user inputs the height to be 5, the pyramid should be:
# *
# **
# ***
# ****
# *****

pyramid_height = int(input("Emter the height of the pyramid: "))

print("Here's our pyramid:")

for i in range(pyramid_height):
  # Multiplying a string by an integer just repeats it that many times.
  print("*" * (i+1))
````

````
# Redo Pyramid with nested for-loop

pyramid_height = int(input("Emter the height of the pyramid: "))

print("Here's our pyramid:")

for i in range(pyramid_height):
  for j in range(i+1):
    print("*", end='') # print * for value of j going from 0,1,..,i

  print() # after printing required number of * in a given level, we need to move to new line for next level (i)
````

````
# Problem: Let's say we want to print out all numbers from 1-100 that are divisible
# by a number given by the user.

given_num = int(input("Please enter a number between 0-100: "))

for i in range(1, 101): # Get numbers from 1-100
  if (i % given_num) == 0:
    print(f"{i} is divisble by our number {given_num}")

print("Done with our loop.")
````

### Lists

These represent a linear series of items / values. The values in a list are _oredered_ and can be accessed via an _index_. Lists are iterable and thus are prime use-cases for _for-loops_.

These have various uses such as storing a list of classes for a class scheduler, a list of transactions for a e-wallet application, a list of items in a shopping cart, and more. 

````
# Problem: We have a list of items in a shopping cart, and we want to print them
# out one after the other.

shopping_cart_items = ["Apple", "Oranges", "Flour", "Sugar", "Salt", "Pepper"]

for item in shopping_cart_items:
  print(item)
````

#### Indexing

To access a specific item in a list, you can use its _index_. An index is just the position at which an item exists in a list. The only thing to note is that the _first_ item in a list is at index _0_. 

Thus for a list that has 10 items in it, the item indices range between 0-9.

Also, you can get the length of a list using the `len(<list_variable/value>)` function.

````
# Here's a few examples of how indexing works:

shopping_cart_items = ["Apple", "Oranges", "Flour", "Sugar", "Salt", "Pepper"]

# 'Apple' is the first item, and will be at index 0
print("The FIRST item in the list is", shopping_cart_items[0])

# 'Flour' is the third item and is at index 2
print("The THIRD item in the list is", shopping_cart_items[2])

# The length of the list is 6, thus 'Pepper', the last item, should be at index 5
print("The LAST item in the list is", shopping_cart_items[5])

# You can access the last element using the `len` function too
print("The LAST item in the list is", shopping_cart_items[len(shopping_cart_items)-1])

# Python also allows negative indices (other languages might not allow them).
# These represent positions from the END of the list
print("The LAST item in the list is", shopping_cart_items[-1])
````

````
# Problem: We can recreate the shopping cart examples above using indices rather
# than the items themselves.

shopping_cart_items = ["Apple", "Oranges", "Flour", "Sugar", "Salt", "Pepper"]

for i in range(len(shopping_cart_items)):
  print(shopping_cart_items[i])
````

### Break and Continue

Sometimes we might want to skip a few iterations / repetitions based on a condition or stop the whole loop when a condition is met. We use `continue` and `break` for this:

* `continue`: Skips forward to the next iteration of the loop, skipping over the statements in the loop below the `continue`

````
# Problem: Let's say we want to print all odd numbers between 2-100, we can do that
# using continue.

for i in range(2, 101):
  if (i % 2) == 0:
    continue
  print("")

# There's another way to do this where we only print when the number is odd, but
# this is another way of doing it!
````


* `break`: 'Breaks'/exits the loop, skipping over the statements in the loop below the `break`

````
# Problem: We want to protect one of the programs with a passcode, and unless the
# user enters the correct passcode, we don't want to let them in. 

# Let's try implementing this using a while loop and break

passcode = "HOWDY"

while True:
  user_input = input("Please enter the passcode: ")
  
  if user_input == passcode:
    break

print("You've passed our security check!")
````
